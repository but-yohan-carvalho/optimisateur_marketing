```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("data/bank.csv")
print (df.head())
```

```python
taux = df['deposit'].value_counts(normalize=True) * 100

plt.figure(figsize=(6, 5))
plt.bar(taux.index, taux.values, color=['#e76f51', '#2a9d8f'], edgecolor='black', width=0.6)

plt.title("Taux global de souscription (deposit)", fontsize=13, fontweight='bold', pad=15)
plt.ylabel("Pourcentage (%)", fontsize=11)
plt.ylim(0, 100) # Force l'axe Y à aller jusqu'à 100%

# Ajouter les valeurs textuelles au-dessus des barres
for i, val in enumerate(taux.values):
    plt.text(i, val + 2, f"{val:.1f}%", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
